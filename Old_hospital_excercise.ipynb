```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/hospital-costs.csv")
```

```python
df.info()
```

```python
df.columns
```

```python
df['Total Charges'] = df["Discharges"] * df["Mean Charge"] # Total charges 
```

```python
df["Total Costs"] = df["Discharges"] * df["Mean Cost"] # Total Cost
```

```python
df["Markup"] = df["Total Charges"]/ df["Total Costs"] #Markup
```

```python
Markups = df.sort_values(by='Markup', ascending=False)
Markups = df[['Facility Name', 'Total Charges', 'Total Costs', 'Markup']]
Markups.head(5)
```

```python
Markups = df.sort_values(by='Markup', ascending=True, )

Markups = Markups[['Facility Name', 'Total Charges', 'Total Costs','Markup']]
```

```python
Markups.head(5) # Top five markup highest cost for hospital
```

```python
Markups.tail(5) #Top five lowest mark up for hospital
```

```python
(df[['APR DRG Description', 'Discharges']]  # Top 10 most reasons people visit hospital
    .groupby('APR DRG Description')
    .sum()
    .sort_values(by='Discharges', ascending=False)
    .head(10))
```

```python
net = df[['Facility Name', 'Total Charges', 'Total Costs']]
    
```

```python
net = net.groupby('Facility Name').sum()
```
